```python
def taylorx(r,t0,t,n):
  h = t - t0
  #dois primeiros termos
  tay = r*(t0 - np.sin(t0)) + (r - r*np.cos(t0))*h
  #outros termos
  for i in range(2,n+1):
    if i%4 == 1:
      d =  - r*np.cos(t0)
    if i%4 == 2:
      d =  r*np.sin(t0)
    if i%4 == 3:
      d = r*np.cos(t0)
    if i%4 == 0:
      d =  - r*np.sin(t0)
    tay = tay + (d*h**i)/np.math.factorial(i)
  return tay
```

```python
def taylory(r,t0,t,n):
  h = t - t0
  #dois primeiros termos
  tay = r*(t0 - np.sin(t0)) + (r*np.sin(t0))*h
  #outros termos
  for i in range(2,n+1):
    if i%4 == 1:
      d =   r*np.sin(t0)
    if i%4 == 2:
      d =  r*np.cos(t0)
    if i%4 == 3:
      d = - r*np.sin(t0)
    if i%4 == 0:
      d =  - r*np.cos(t0)
    tay = tay + (d*h**i)/np.math.factorial(i)
  return tay
```
